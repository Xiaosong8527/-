# java数据类型的定义
- 任何的程序严格来讲都属于一个数据的处理游戏；
- 对于数据的保存就必须有严格的限制，这些限制就体现在数据类型的划分上
- 不同的数据类型就可以保存不同的数据内容


## java数据类型划分
### java主要分为两大数据类型
### （基本数据类型）
### （引用数据类型）
- 基本数据类型不牵扯到内存分配问题
- 引用数据类型需要开发者为其分配空间，而后进行关系的匹配
- 基本数据类型一共分为八种，引用数据类型一共分为三种

### 基本数据类型
- 基本数据类型
    - 数值型：
        - 整型：byte;short;int;long  默认值：0
        - 浮点型：float;double     默认值：0.0
    - 字符型：char    默认值：'\u0000'
    - 布尔型：boolean  默认值：false
- 引用数据类型：
    - 数组；类；接口； 默认值：null

##  选择数据类型原则
- 如果要是表示整数就使用  int
- 如果表示小数使用 double
- 如果要描述日期时间数字或者表示文件或者内存大小使用 long；
- 如果要实现内容传递或者是编码转换使用 byte；
- 如果要想实现逻辑的控制，可以使用 boolean描述
- 如果要想描述中文，使用char可以避免乱码问题
- 如果按照保存范围：byte<int<long<double

## 整型
- 任何一个数字常量（例如：300；100）都属于int数据类型
- 即在java中所有设置的整数内容默认情况下都是int型数据
        
        public class className{
            public static void main(String args[]){
             // 为变量设置内容使用如下格式：
             // 数据类型 变量名称 = 常量
             int num = 10; // 10是常量，常量的默认类型是int
             int result = num * 2; // 利用num变量的内容乘以2
                           // 并赋值给 result
             System.out.println(result);
            }
        }
- 定义完变量后可以对变量进行四则运算；
- 实际上变量与常量最大的区别只有一个：常量的内容是固定的，变量的内容是可变的
- 所有的变量名称在同一代码块中只允许声明一次
        
        int num = 10;
        num = num * 2; 
        // 第二次使用变量 num 时不需要声明
        

- 每一种数据类型都有其对应数据类型的保存范围
    - 如果最大值继续增加就会变成最小值，随后一直向其次的最小值进行循环
    - 最小值减1就变为了最大值，随后一直向其次的最大值进行循环
    - 因为任何的数据运算都是按照二进制进行的，第一位时符号位，其余的31位为数据位
        - 此种现象称为数据的溢出，如果要解决这种溢出问题，就只有通过扩大数据范围的方式来实现，比int更大的时long数据类型
- 在程序中，数据类型的转换有以下规律
    - 数据范围小的数据与数据范围大的数据进行数学计算，自动向大的数据范围类型进行转换
    - 数据范围大的数据变为数据范围小的数据，必须采用强制转换
    - 以上的操作都是针对变量的
    - 如果常量进行强制转换，有两种方法
        - 常量标记（L;l) 使用“（数据类型）”

        public class className{
            public static void main(String args[]){
                int max = Integer.MAX_VALUE; //取出int最大值
                int min = Integer.MIN_VALUE; //取出int最小值
                System.out.println(max);   //2147483647
                System.outprintln(min);    //-2147483648
                //int 变量 + 或 - long型常量 = long型数据
                System.out.println(max + 1L);  //最大值加1
                System.out.println(min - (long)1); //最小值减1
                //long变量 +- int型常量 = long型数据
                System.out.print((long)min - 2); 
            }
        }
        
- 只要写的代码属于正常可以使用的代码，基本上数据溢出的情况除了极端情况下基本不会出现

## 大范围数据类型向小范围数据类型进行强制转换

        public class className{
            public static void main(String args[]){
                long num = 1000; //1000常量是int类型，使用long接收
                int x = (int) num; // 把long转换为int
                System.out.println(x);
            }
        }
- 如果long变为int时所保存的数据超过了int的数据范围时，会出现数据溢出的现象
- 在实际使用中尽量不要使用long， 除非表示日期时间，文件大小或内存时使用

## 整型 byte
- byte取值范围是 -128 ~ 127之间
- 1.byte转换问题

        public class classname{
            public static void main(String args[]){
                int num = 130; //此数据超过byte定义
                byte x = (byte)num   //由int变为byte类型
            }
        }
- 由于byte使用的特殊性，java对其有了一些很好的改善。

        public class classname{
            public static void main(String args[]){
                byte num = 100; 
                System.out.println(num);
            }
        }
- 虽然任何一个整数都属于int型，但是在java中，如果发现使用的数据变量类型为lyte并且设置的数据范围在byte数据范围内，那么就会自动帮助用户实现数据类型的转换。反之如果超过了，则依然会自动以int型为主。
- 所有的变量在使用的时候一定不要去相信默认值，都设置具体内容；
- 如果在方法里面定义的默认值是不起任何作用的；
        
        public class classname{
            public static void main(String args[]){
                int num; //没有默认值
                num = 0; //在使用此变量之前设置内容
                System.out.println(num);
            }
        }
    - 以上内容使先定义了一个变量，但是此变量没有设置内容，而后设置了num变量的内容，最后再使用此变量，可是以上的代码如果再JDK1.4及以前的版本是不可能编译通过的。
    - 最好的方法是在定义变量的时候直接设置好默认值
    

           public class classname{
            public static void main(String args[]){
                int num = 0；
                
                System.out.println(num);
            }
           }
     - 以上代码是java中变量赋值的标准写法

- 自己编写代码时，90%都去考虑使用int数据类型
        

# 浮点数 
- 在java中只要是小数，对应的数据类型就时double型数据（double是保存范围最广的数据类型）

        public class classname{
            public static void main(String args[]){
            //10.2是一个小数，所以属于double数据类型
            double num = 10.2;
            //double型 * int型（转化为double，2.0）= double型
            System.out.print(num*2);
            }
        }
- 由于默认的小数类型就是double，所以如果使用了float表示需要double型变为float型，需要采用强制转换

        public class classname{
            public static void main(String args[]){
                float f1 = 10.2F;
             或者 float f2 = (float)10.2;
             System.out.println(f1*f2);
            
            }
        }
    - 最终发现结果出现问题，变为了 104.03999，这个问题如果要追溯起来从 JDK1.0就出现这个bug，解决不了，只能通过后期的处理来完成。这个bug不会影响开发。

- 额外：
    - 实际上最早开发的时候，考虑到内存问题，往往能使用float就不适用double。例如：J2ME开发，由于内存苛刻，往往会压缩数据范围以节约空间。随着硬件成本降低，时候使用double和float区别不大，所以直接使用double数据。
    - 需要注意的是，所有的数据类型只有double才可以保存小数

          public class classname{
            public static void main(String args[]){
                int x = 9;
                int y = 5;
                    System.out.println(x / y);
           }
           }
    
     - 实际上x / y 的结果为1.8，但是编译结果为1，因为所有的小数位背忽略了，
     - 如果此时要想正确的得出所要的结果，则可以将一种一个整型变为浮点型
     
           public class classname{
            public static void main(String args[]){
                int x = 9;
                int y = 5;
                    System.out.println(x / (double)y);
           }
           }
     - 在开发中一定要考虑到整型不保留小数位的问题

# 字符型 char
- byte，是属于字节，按照传统的概念，一个字符 = 两个字节；
- 对于字符，除了与字节有些关系外，最主要的关系在于与int型变量的转换
    - 在计算机中一切都是以编码的形式出现的；java使用的是十六进制的UNICOD编码
    - 此类编码可以保存任意的一个文字，但是这个编码在设计的过程中，考虑到与其他语言的结合问题(C/C++)。那么在此编码里面包含了ASCII码的部分编码。

- 在程序中使用单引号声明的内容称为字符。每一个单引号里面只能保存一位字符

        定义字符
        public class classname{
            public static void main(String args[]){
                 char c = 'A';
                 //字符可以和int型互相转换
                 //转换后以编码的形式出现
                 int num = c;
                 System.out.println(c);
                 System.out.println(num); //此条代码输出结果为 65
          }
        }    

    - 'A'（65） ~ 'Z'（90）
    - 'a'（97） ~ 'z'（122）
    - '0'（48） ~ '9'（57）
        - 其中 “A”的编码值要小于“a”编码值32，那么可以利用简单的数学计算实现大小写转换
-
    
        public class classname{
            public static void main(String args[]){
                 char c = 'A'; //是大写字母
                 int num = c; // 需要将字符变为int型才可以使用加法计算
                 num = num + 32; // 变为小写字母的编码
                 c = (char)num; //将int变为char型,需要进行强制转换
                 System.out.println(c);
           }
         }
- 同一个英文字母字符，大小写之间的编码值永远相差 32                 
- 传统变成语言之中，字符里面只能够保存一些英文字母的标记，在java之中，由于使用了UNICOD编码 ，这种十六进制的编码可以保存任意的文字，可以设置一个中文字符，一个汉字就是一个字符。     
    - 因为目前的中文保存方便，所以在处理断句的时候很好用。
    - 只有在处理中文的时候，字符数据才有那么一点帮助，其他时候基本用不到字符数据
                 
            
            

# 布尔型 boolean
- 布尔型是一种逻辑结果，主要保存两种数据：true和false
- 这类数据主要用于程序的逻辑使用上
    - 应该和python中的布尔类型一样吧 。。。。。
    - python中是 True和False
    - java中全是小写
    - java中布尔型的数据类型默认值是 false
- 
        
        public class classname{
            public static void main(String args[]){
                boolean flag = true；
                if (flag){ //if（布尔值） {大括号中为满足条件的操作}
                // 可以理解为python中if语句中的条件表达式
                    System.out.println("Hello World!")
            }
          }
        }
- 如果 if (flag) 变为if (!flag)  则为flase  那么条件表达式则不成立 
    - 感叹号理解为python中 ==表示等于 != 表示不等于
- 由于许多编程语言在设计时并没有考虑到布尔型的问题，那么就使用了数字0表示flase
- 非0数字表示ture
- 这样的设计对于代码开发比较混乱，JAVA里面不允许使用0或1来填充布尔型的变量内容

# String型数据
- java中表示字符串的数据类型 String
    - 再python中字符串可以用单引号或者双引号表示
    - 在java中，单引号表示 字符 ，双引号才表示字符串
       
- 只要是项目开发，百分之百会使用String；
- 但是与其他几种基本数据类型相比String属于引用数据类型（它）属于一个类
- 在java里面只要是类名称，每一个单词首字母都是大写的（String）
- 但是这个类的使用比较特殊
- String标识的是一个字符串，需要使用双引号声明其内容
    - 一定不能像python中使用单引号表示字符串
- 在String中，也可以使用 + 号进行字符串链接操作，与python相似

        public class ClassName{
            public static void main(String args[]){
            // 数据类型 变量名 = “内容”
                 String str = "Hello World!";
                     System.out.println(str);
            }
        
        }

- 字符串连接
        
        public class ClassName{
            public static void main(String args[]){
            // 数据类型 变量名 = “内容”
                 String str = "Hello ";
                 str = str + "World!"; // 连接字符串
                 str += "World！"; // 这种方式也表示链接字符串
                 // 与python中 += 相似
                 // str += 1 = (str + 1)
                     System.out.println(str);
            }
        }
- 如果使用+ 号对String数据类型与其他数据类型进行链接，则按照之前的
- 任何数据类型都向范围大的数据类型进行转换
- 如果是 int + double，那么int先变为double，再进行加法运算
- 如果int 遇见 String ，则可以理解为所有数据类型如果遇见了String的 + 操作（只有+）的情况
- 那么所有的数据类型都先变为 String数据，然后再进行字符串相加操作
        
        public class ClassName{
            public static void main(String args[]){
                 int numA = 100;
                 double numB = 99.0;
                 String str = "字符串" + (numA + numB);
                     System.out.pringln(str);
              // 与python相同，使用括号定义运算优先级
              // 括号在所有运算符号中具有最高优先等级
            }
            
        }
- 先执行括号内的加法运算，而后再将结果（结果为String类型）与字符串进行连接
- 在java里面也执行多种转义字符的使用。
    - 与python相同的是，转义字符的 \  同样具有特殊意义
    - \n 换行
    - \t 制表符
    - \\ 表示一个斜线
    - \' 单引号
    - \" 双引号
- 
        
        public class ClassName{
            public static void main(String args[]){
                String str = "Hello \"World\""; //结果为 Hello "World"  
            }  
        }
- 这些转义字符在学子中考虑到格式显示会经常出现，但是在开发过程中使用的情况是比较少的

# 总结 （一般来说）
- 整数用 int
- 小数用 double
- 逻辑用 boolean
- long和byte和char再处理数据时会使用
- 数据类型的转换永远都是小范围的数据类型自动向大范围的数据类型转换
- 如果将大范围数据类型变为小范围数据类型则要考虑到强制转换
- int 和 char 之间的互相转换是以编码为主，大小写字母之间差了32长度
- String是一个特殊的类，用于描述字符串，字符串使用双引号声明，利用 + 号进行字符串拼接
